INTEGRALS

```python
import sympy as smp
from sympy import *
```

```python
#define symbols
t,s=smp.symbols('t s', positive=True)
a,l0,l1,m1,m2 = smp.symbols('a,l0,l1,m1,m2', positive=True)
```

```python
#DEFINE A FUNCTION
P01=(l0/a)*(1-exp(-a*t))
sum_1=l1*P01
```

```python
#Substitute a variable from a given fuction: sum_1.subs(t,s)
#sum_1.subs(t,s)

I=exp(-m2*(t-s))*sum_1.subs(t,s) #Integrand
R=smp.integrate(I,(s,0,t))
```

```python
simplify(R)
```

```python
expand(R) #Let's take only the solution when a distinct m2
```
